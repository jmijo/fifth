### 3. 작사작곡가

```python
import re

def extract_roles(text, role):
    pattern = rf'(.+?)\n{role}'
    matches = re.findall(pattern, text)
    return matches 
```

```python
melon['작사가'] = melon['작곡작사가'].apply(lambda x: extract_roles(x, '작사'))
melon['작곡가'] = melon['작곡작사가'].apply(lambda x: extract_roles(x, '작곡'))
melon['편곡가'] = melon['작곡작사가'].apply(lambda x: extract_roles(x, '편곡'))
```
