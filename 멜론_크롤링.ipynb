```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time
```

### 연도, 순위, 노래, 가수, 가사, 장르, 앨범소개 크롤링 코드

```python
# 2009 (1)
browser = webdriver.Chrome()
browser.maximize_window()
browser.get('https://www.melon.com/chart/index.htm')

browser.find_element(By.XPATH,'//*[@id="gnb_menu"]/ul[1]/li[1]/div/div/button').click() # 차트 파인더 클릭
time.sleep(0.1)

data2009 = []

browser.find_element(By.XPATH,'//*[@id="d_chart_search"]/div/h4[3]').click() # 연도 차트 선택
browser.find_element(By.XPATH,'//*[@id="d_chart_search"]/div/div/div[1]/div[1]/ul/li[3]').click() # 연대 선택 - 2000년대
time.sleep(0.1)

browser.find_element(By.XPATH, '//*[@id="d_chart_search"]/div/div/div[2]/div[1]/ul/li[1]').click() # 연도 선택 - 2009년
time.sleep(0.1)
browser.find_element(By.XPATH,'//*[@id="d_chart_search"]/div/div/div[5]/div[1]/ul/li[1]').click() # 장르 선택
browser.find_element(By.XPATH,'//*[@id="d_srch_form"]/div[2]/button').click() # 검색 선택
time.sleep(0.2)
year = browser.find_element(By.XPATH,'//*[@id="serch_cnt"]/span[1]').text
containers=browser.find_elements(By.XPATH, '//*[@id="lst50"]') 
               
for elem in containers:
                try:
                    순위 = elem.find_element(By.CLASS_NAME, 'rank').text
                except:
                    순위 = ' '
                try:
                    노래제목 = elem.find_element(By.CLASS_NAME, 'rank01').text
                except:
                    노래제목 = ' '           
                try:
                    가수 = elem.find_element(By.CLASS_NAME, 'rank02').text
                except:
                    가수 = ' '
    
                elem.find_element(By.CLASS_NAME,'btn_icon_detail').click()
                time.sleep(0.1)
                try:
                    browser.find_element(By.XPATH, '//*[@id="lyricArea"]/button').click()     # 가사 펼치기
                    time.sleep(0.1)
                    가사 = browser.find_element(By.XPATH, '//*[@id="d_video_summary"]').text         # 가사 가져오기
                except:
                    가사 = ' '
                try: 
                    작곡작사가 = browser.find_element(By.CLASS_NAME, 'list_person').text
                except:
                    작곡작사가 = ' '
                time.sleep(0.1)
                try: 
                    장르 = browser.find_element(By.XPATH, '//*[@id="downloadfrm"]/div/div/div[2]/div[2]/dl/dd[3]').text
                except:
                    장르 = ' '                       
                browser.back()
                time.sleep(0.01)
        
                elem.find_element(By.CLASS_NAME,'rank03').click()
                try: 
                    browser.find_element(By.XPATH,'//*[@id="conts"]/div[4]/div[2]/button').click() # 앨범 더보기 클릭
                    time.sleep(0.1)
                    앨범정보 = browser.find_element(By.CLASS_NAME, 'dtl_albuminfo').text
                except:
                    앨범정보 = ' '
                browser.back()
                time.sleep(0.01)
    
                data2009.append([year, 순위, 노래제목, 가수, 가사, 작곡작사가, 장르, 앨범정보])
```

### 가수 유형, 인원수 크롤링 코드

```python
# 2024년

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

browser = webdriver.Chrome()
browser.maximize_window()
browser.get('https://www.melon.com/chart/index.htm')

album = []
browser.find_element(By.XPATH,'//*[@id="gnb_menu"]/ul[1]/li[1]/div/div/button').click() # 차트 파인더 클릭
browser.find_element(By.XPATH,'//*[@id="d_chart_search"]/div/h4[3]').click() # 연도 차트 선택

page = 50 
browser.find_element(By.XPATH,'//*[@id="d_chart_search"]/div/div/div[1]/div[1]/ul/li[1]').click() # 연대 선택 - 2020년대
time.sleep(0.1)

browser.find_element(By.XPATH, '//*[@id="d_chart_search"]/div/div/div[2]/div[1]/ul/li[1]').click() # 연도 선택 - 2024년
time.sleep(0.1)

browser.find_element(By.XPATH,'//*[@id="d_chart_search"]/div/div/div[5]/div[1]/ul/li[1]').click() # 장르 선택
browser.find_element(By.XPATH,'//*[@id="d_srch_form"]/div[2]/button').click() # 검색 선택

wait = WebDriverWait(browser, 10)
containers = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, 'rank02')))

time.sleep(0.1)
for elem in containers:
    try:
        elem.click()
        time.sleep(0.1)
        browser.find_element(By.XPATH,'//*[@id="conts"]/div[3]/ul/li[2]').click() # 상세정보 클릭
        time.sleep(0.1)
        try: 
                result = browser.find_element(By.XPATH, '//*[@id="conts"]/div[6]/dl').text.split('\n')
                index = result.index('유형')
                value = result[index+1]
        except:
                value = ' '
        try: 
                if value.split(' |')[0] == '솔로':
                    member = 1
                else:
                    member = len(browser.find_element(By.CLASS_NAME, 'wrap_atistname').text.split(','))
        except:
                member = ' '
        album.append([value, member])
        browser.back()
        browser.back()
        if len(album) == page:
            browser.find_element(By.XPATH,'//*[@id="frm"]/div[2]/span/a').click() # 51~100위 순위 페이지 클릭
            time.sleep(0.2)
        else:
            pass
    except:
        album.append([' ', ' '])
```

```python
유형 = pd.DataFrame(album)  
유형
```

```python
유형.to_csv('album_2024.csv',encoding = 'utf-8-sig')
```
